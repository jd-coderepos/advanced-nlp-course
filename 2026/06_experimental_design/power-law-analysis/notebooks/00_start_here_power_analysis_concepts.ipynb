## Core inputs

Power analysis usually needs more than the final scores.

For example, the pair below is not enough by itself:

| Model A | Model B | Difference |
|---:|---:|---:|
| 90% | 91% | +1% |

You also need at least:

- sample size: how many test instances or human-rated items?
- test choice: McNemar, paired t-test, bootstrap/randomization, mixed-effects model, etc.
- variability/agreement: do the methods fail on mostly the same items, or on different items?
- alpha: significance threshold, often 0.05
- target power: often 0.80


```python
# A small illustration of why sample size matters.
# Students can edit these values.
comparison = pd.DataFrame({
    "Test examples": [100, 10_000],
    "Model A": [0.90, 0.90],
    "Model B": [0.91, 0.91],
})
comparison["Difference"] = comparison["Model B"] - comparison["Model A"]
comparison
```

```python
# Approximate standard error for a single accuracy estimate.
# This is not the full paired comparison, but it gives useful intuition.
def accuracy_standard_error(p, n):
    return np.sqrt(p * (1 - p) / n)

comparison["Approx. SE of 90% accuracy"] = [accuracy_standard_error(0.90, n) for n in comparison["Test examples"]]
comparison["Approx. 95% margin"] = 1.96 * comparison["Approx. SE of 90% accuracy"]
comparison
```
