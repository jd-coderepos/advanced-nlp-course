# 04 — Paired Student's t-Test for Numeric Scores

Use a paired t-test when each unit has a numeric score from both methods.

Examples:

- the same generated item rated for Model A and Model B;
- the same random seed used for two systems;
- the same dataset split evaluated under two conditions.

```python
from pathlib import Path
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Make src/ importable when running from the notebooks folder.
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

from significance_utils import *

DATA_DIR = PROJECT_ROOT / 'data' / 'dummy'
ALPHA = 0.05
```

## Example A: human ratings by item

```python
ALPHA = 0.05
ratings = pd.read_csv(DATA_DIR / 'human_ratings.csv')
ratings.head()
```

```python
result = paired_t_test(
    scores_a=ratings['model_a_mean_rating'],
    scores_b=ratings['model_b_mean_rating'],
)

pd.Series(result)
```

```python
print(interpret_p_value(result['p_value'], alpha=ALPHA))
```

```python
diffs = ratings['model_b_mean_rating'] - ratings['model_a_mean_rating']
ax = diffs.plot(kind='hist', bins=20)
ax.set_xlabel('Per-item rating difference: Model B - Model A')
ax.set_title('Paired rating differences')
plt.show()
```

## Example B: repeated runs / random seeds

```python
runs = pd.read_csv(DATA_DIR / 'repeated_runs.csv')
runs
```

```python
result_runs = paired_t_test(
    scores_a=runs['model_a_f1'],
    scores_b=runs['model_b_f1'],
)

pd.Series(result_runs)
```

```python
print(interpret_p_value(result_runs['p_value'], alpha=ALPHA))
```
