# 00 — Start Here: Statistical Significance Tests

This exercise is about **statistical significance testing** for NLP experiments.

The basic question is:

> Given the dataset I already evaluated on, is the observed difference between two methods large enough to reject the null hypothesis of no real difference?

This is different from power analysis. Power analysis asks whether the dataset is large enough before or during experimental design. Significance testing analyses observed results.

## Four tests in this exercise

| Test | Typical use case | Data shape |
|---|---|---|
| McNemar's test | paired classification accuracy | same examples, two predicted labels |
| Paired bootstrap test | per-instance metric scores | same examples, two numeric scores |
| Paired randomization test | NLP system outputs, MT-style evaluation | same examples, two outputs, one reference |
| Paired t-test | ratings or repeated runs | same items/seeds, two numeric scores |
