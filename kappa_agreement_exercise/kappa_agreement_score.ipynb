```python
import pandas as pd

# Load the CSV file
df = pd.read_csv("annotation_data.csv")

# Rename for convenience
labels1 = df["Annotator A"]
labels2 = df["Annotator B"]

df
```

Step 1: Confusion Matrix

```python
conf_matrix = pd.crosstab(labels1, labels2,
                          rownames=["Annotator A"], colnames=["Annotator B"])

conf_matrix
```

Step 2: Observed Agreement

```python
# Total agreements on the diagonal
agree = sum(conf_matrix.iloc[i, i] for i in range(min(conf_matrix.shape)))
total = len(df)

P_o = agree / total
print(f"Observed Agreement (P_o): {P_o:.3f}")
```

Step 3: Expected Agreement

```python
# Label distributions
a1_dist = labels1.value_counts(normalize=True)
a2_dist = labels2.value_counts(normalize=True)

# Expected agreement
all_labels = sorted(set(a1_dist.index).union(set(a2_dist.index)))
P_e = sum(a1_dist.get(label, 0) * a2_dist.get(label, 0) for label in all_labels)

print(f"Expected Agreement (P_e): {P_e:.3f}")
```

Step 4: Cohen's Kappa

```python
kappa = (P_o - P_e) / (1 - P_e)
print(f"Cohen's Kappa: {kappa:.3f}")
```

Step 5: Interpretation

```python
def interpret_kappa(kappa):
    if kappa < 0:
        return "Less than chance agreement"
    elif kappa <= 0.20:
        return "Slight agreement"
    elif kappa <= 0.40:
        return "Fair agreement"
    elif kappa <= 0.60:
        return "Moderate agreement"
    elif kappa <= 0.80:
        return "Substantial agreement"
    else:
        return "Almost perfect agreement"

print("Interpretation:", interpret_kappa(kappa))
```
